# Actividades - Python

  Desarrolar el código en python que permita realizar las siguientes instrucciones.
*   Use las estrategías de consulta apropiadas para:  
    <b>i</b> Buscar registros duplicados y eliminarlos. 
    
    <b>ii</b> Chequear el formato de los datos (numérico en las columnas en que deba haber números, texto en las columnas correspondientes, etc.). 
    
    <b>iii</b> Separar la columna "duration" en dos: en "duration" deben quedar solo los números, mientras que la que quede con los textos deben llamarla "tiempo_temporadas".
    
    <b>vi.</b> Construir un DF  que muestre de manera descendente los 15 países que tienen mayor cantidad total de títulos en la plataforma, desglosando por tipo de contenido.
    
    <b>v.</b> Mostrar las películas clasificadas como **drama** por netflix, que hayan ganado el Oscar entre 2010 y 2020 y que estén en esta plataforma. 
    
    <b>vi.</b> Construir un DF que muestre el nombre de las películas nominadas, el año de la ceremonia correspondiente, la cantidad de nominaciones y la cantidad de oscar ganados. Agregar una columna que sea el resultado de calcular la relación entre estatuillas ganadas y nominaciones totales. Organizar la tabla con mayor o igual a 10 nominaciones y de mayor a menor por esa de la relación calculada, limitada sólo a las 20 películas de mejor perfomance en los oscar. 

```
import pandas as pd
from IPython.display import display

# i

print("Respuesta i - \'Buscar registros duplicados y eliminarlos\':")
#Abro el archivo como un Dataframe
set_total = pd.read_csv('/work/Data/set_total .csv')
# Verifico la cantidad de registros duplicados
print('Hay', set_total.duplicated().sum(), 'registros duplicados')
print(set_total.shape)
# Elimino registros duplicados
set_total.drop_duplicates(inplace=True)
# Verifico que se hayan eliminado
print(set_total.shape)
print()

# ii

print("Respuesta ii - \'Chequear el formato de los datos\':")
print(f'Información de los tipos de datos de set_total: ')
# Reviso los tipos de datos
print(set_total.dtypes)
print()

# Creo una función que me permita cambiar los tipos de datos
def cambio_type(df, lista, tipo):
    for columna_nombre in lista:
        if tipo == 'int' or tipo == 'float' :
            df[columna_nombre].fillna(0, inplace=True)
        df[columna_nombre] = df[columna_nombre].astype(tipo)
    return df

# Convierto las siguientes columnas a int
lista1 = ['id_content', 'release_year', 'year_ceremony', 'winner']
cambio_type(set_total, lista1, 'int')

# Covierto las siguientes columnas a str
lista2=['title_content', 'cast', 'director', 'country', 'listed_in', 'language', 'title_oscar', 'category', 'name', 'directed_by', 'starring']
cambio_type(set_total, lista2, 'string')

# Covierto las siguientes columnas a category
lista3=['type', 'rating']
cambio_type(set_total, lista3, 'category')

print('Se reemplazaron los valores nulos por cero, y se adecuaron los tipos de datos')
print(f'Información de los tipos de datos de set_total con las modificaciones: ')
print()
print(set_total.dtypes)
print()

# iii

print("Respuesta iii - \'Separar la columna duration en dos\':")
# Realizo un split para separar los datos en dos columnas
set_total[['duration', 'tiempo_temporadas']] = set_total['duration'].str.split(' ', 1, expand=True)
print(set_total[['duration', 'tiempo_temporadas']].head(10))
print()

# iv

print("Respuesta iv - \'Construir un DF  que muestre de manera descendente los 15 países que tienen mayor cantidad total de títulos en la plataforma, desglosando por tipo de contenido\':")
print()
# Separo las columnas necesarias en un nuevo df
df_iv = set_total[[ 'id_content','country','type']]
# Elimino datos nulos
df_iv = df_iv.dropna()
# Elimno datos con coma en la columna country, para evitar contar las colaboraciones
df_iv = df_iv.loc[~df_iv['country'].str.contains(',') ]
# Realizo un groupby con country y type que muestre la cantidad de contenido por país y tipo de contenido
df_iv = df_iv.groupby(['country','type']).count().sort_values(['id_content','country'], ascending=False)

display(df_iv.head(15))
print()

# v

print("Respuesta v - \'Mostrar las películas clasificadas como drama por netflix, que hayan ganado el Oscar entre 2010 y 2020 y que estén en esta plataforma.\':")
print()
# Preparo un df con las columnas y limitaciones necesarias
df_v = set_total.loc[(set_total['listed_in'].str.contains('Drama')) 
                      & (set_total['winner']==1) 
                      & (set_total['year_ceremony']>= 2010) 
                      & (set_total['year_ceremony']<=2020)]
display(df_v.head(5))

# vi

print("Respuesta vi - \'Construir un DF que muestre el nombre de las películas nominadas, el año de la ceremonia correspondiente, la cantidad de nominaciones y la cantidad de oscar ganados. Agregar una columna que sea el resultado de calcular la relación entre estatuillas ganadas y nominaciones totales. Organizar la tabla con mayor o igual a 10 nominaciones y de mayor a menor por esa de la relación calculada, limitada sólo a las 20 películas de mejor perfomance en los oscar.\':")
print()

# Creo un df con las columnas que necesito
df_vi = set_total[['title_oscar','year_ceremony','winner']].reset_index(drop= True)
# Hago un groupby con la cantidad de nomicaciones
df_vi_nom = df_vi.groupby(['title_oscar','year_ceremony']).count()
# Hago un groupby con la cantidad de premios
df_vi_win = df_vi.groupby(['title_oscar','year_ceremony']).sum('winner')
# Nombro de manera más adecuadas las columnas correspondientes
df_vi_nom.columns = ['Nominaciones']
df_vi_win.columns = ['Oscars ganados']
# Combino ambos agrupamientos en el df final
df_vi_final = pd.merge(df_vi_nom, df_vi_win, on=['title_oscar','year_ceremony'], how='inner')
# Agrego la columna de ratio entre oscar ganados y nominaciones
df_vi_final['Ratio'] = (df_vi_final['Oscars ganados'] / df_vi_final['Nominaciones']).round(2)
# Aparecerán solo las películas con más de 10 nominaciones
df_vi_final = df_vi_final[df_vi_final['Nominaciones']>=10]
# Ordeno las peliculas por ratio en orden descendente
df_vi_final.sort_values(by='Ratio', ascending=False, inplace=True)
# Limito a las primeras 20 peliculas por ratio
display(df_vi_final.head(20))
```
